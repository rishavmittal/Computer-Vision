# PIL (Python Image Library)
<p>É uma biblioteca que proporciona manipulações gerais e básico de operações de imagens em Python, como rotação, redimensionar, converter cores, etc. Vamos começar usa-lo daqui pra frente.</p>
![PIL Logo](pil.png)

```python
from PIL import Image
imagem = Image.open('image1.jpg')
```

<p>Ela já retorna a imagem como um objeto. Daí já poderiamos mostra-lá:</p>

```python
imagem
```

<p>Poderiamos usar um método para converter a escla de cor da imagem para escala cinza, por exemplo, dai fazemos:</p>

```python
imagem = Image.open('image1.jpg').convert('L')
imagem
```

<p>Podemos usar o método ** save() ** para salvar a imagem em um novo formato ou forma.</p>

```python
import os
```

```python
rgb_imagem = imagem#.convert('RGB')
rgb_imagem.save('image1.png')
```

<p>Usar o PIL para criar thumbnails é muito fácil, basta usar o método ** thumbnail() **</p>

```python
imagemthumb = imagem
imagemthumb.thumbnail((128, 128))
imagemthumb
```

<p>Também podemos fazer um recorte e uma transposição com os métodos ** crop() ** e ** transpose() **</p>

```python
imagem = Image.open('image1.jpg').convert('L')
caixarecorte = (100,100,400,400)
recorte = imagem.crop(caixarecorte)
recorte
```

```python
imagem = Image.open('image1.jpg').convert('L')
transposicao = imagem
transposicao.transpose(Image.ROTATE_180)
```

```python
transposicao = recorte.transpose(Image.ROTATE_180)
imagem.paste(transposicao, caixarecorte)
imagem
```

<p>Por ultimo, podemos fazer uma rotação ou um redimensionamento com ** rotate() ** e o ** resize() **</p>

```python
saida = imagem.resize((128, 128))
saida
```

```python
saida = imagem.rotate(45)
saida
```
